```python
import pandas as pd
import pickle
from datetime import timedelta
```

```python
DIR = "../"
PATH_TIME_SERIES_DST = DIR + "df_timeseries_en.tsv.gz"
```

```python
timeseries = pd.read_csv(PATH_TIME_SERIES_DST, sep = '\t', compression='gzip')
timeseries['datetime']= pd.to_datetime(timeseries['datetime'])
timeseries['delta_views'] = timeseries['delta_views'].round(decimals = 3)
timeseries['subs'] = timeseries['subs'].round(decimals = 3)
timeseries['delta_subs'] = timeseries['delta_subs'].round(decimals = 3)
timeseries.head()
```

```python
timeseries.datetime = pd.to_datetime(timeseries.datetime - timedelta(hours = 2)).dt.date
timeseries = timeseries[timeseries.datetime > pd.to_datetime("2017-3-1")]
timeseries.head()
```

```python
channel_timeseries = pd.pivot_table(timeseries, values='delta_views', index=['datetime'], columns=['channel'])
channel_timeseries.head()
```

```python
channel_timeseries = channel_timeseries.dropna(thresh=len(channel_timeseries) - 1, axis=1)
channel_timeseries.head()
```

```python
nan_rows = channel_timeseries[channel_timeseries.isna().any(axis=1)]
print(nan_rows.index.values)
channel_timeseries.drop(nan_rows.index,inplace=True)
channel_timeseries.head()
```

```python
sample_set = channel_timeseries.sample(n=5000,axis='columns')
sample_set.head()
```

```python
file = open('clustering', 'wb')
pickle.dump(sample_set, file)
file.close()
```
